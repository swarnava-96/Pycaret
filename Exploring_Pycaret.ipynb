## Dropping Features

```python
pokemon=get_data('pokemon')
```

```python
clf=setup(data=pokemon,target='Legendary',ignore_features=['#','Name'])
```

```python
clf
```

### Numerical to Categorical

```python
hepatitis=get_data('hepatitis')
```

```python
clf=setup(data=hepatitis,target='Class',categorical_features=['AGE'])
```

```python
clf
```

```python
clf=setup(data=pokemon,target='Legendary')
```

```python
clf
```

### Ordinal Encoding

```python
employee=get_data('employee')
```

```python
employee['salary'].value_counts()
```

```python
clf1=setup(data=employee,target='left',ordinal_features={'salary':['low','medium','high']})
```

```python
clf1
```

## Cardinal Encoding

```python
income=get_data('income')
```

```python
income['native-country'].unique()
```

```python
clf=setup(data=income,target='income >50K',high_cardinality_features=['native-country'])
```

```python
clf
```
